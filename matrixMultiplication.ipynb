```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file generated by the C++ code
data = pd.read_csv("matrix_multiplication_performance.csv")

# Print the data to verify it's loaded correctly
print(data)

# Plot the total cycles, compute cycles, load cycles, and utilization
fig, ax = plt.subplots(figsize=(10, 6))

# Plot Total Cycles, Compute Cycles, and Load Cycles
ax.bar(['Total Cycles', 'Compute Cycles', 'Load Cycles'], 
       [data['Total Cycles'][0], data['Compute Cycles'][0], data['Load Cycles'][0]], 
       color=['blue', 'green', 'red'], alpha=0.7)

# Plot Utilization as a percentage
ax2 = ax.twinx()  # Create a second y-axis to plot utilization
ax2.plot(['Utilization'], [data['Utilization (%)'][0]], color='orange', marker='o', markersize=8, linestyle='-', linewidth=2)

# Customize the plot
ax.set_title('Matrix Multiplication Performance')
ax.set_xlabel('Performance Metrics')
ax.set_ylabel('Cycles')
ax2.set_ylabel('Utilization (%)')

# Add labels on the bars for clarity
for i, v in enumerate([data['Total Cycles'][0], data['Compute Cycles'][0], data['Load Cycles'][0]]):
    ax.text(i, v + 0.05 * max([data['Total Cycles'][0], data['Compute Cycles'][0], data['Load Cycles'][0]]), 
            f'{v:.2f}', ha='center', va='bottom', fontsize=10)

# Display the plot
plt.tight_layout()
plt.show()
```
